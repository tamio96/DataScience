```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist
from sklearn.cluster import KMeans
import matplotlib.image as mpimg
import pandas as pd

%matplotlib inline

np.random.seed(123)
```

```python
means = [[2, 2], [5, 6], [1, 8]]
cov = [[1, 0], [0, 1]]
N = 500
K = 3

X0 = np.random.multivariate_normal(means[0], cov, N)
X1 = np.random.multivariate_normal(means[1], cov, N)
X2 = np.random.multivariate_normal(means[2], cov, N)

X = np.concatenate([X0, X1, X2], axis = 0)

original_label = np.asarray([0]*N+[1]*N+[2]*N)
```

```python
def kmeans_display(X, label):
    K = np.amax(label) + 1
    
    plt.scatter(x=X[label==0, 0], y=X[label==0, 1], marker='^', alpha=.8, c='b', label='Kind 1')
    plt.scatter(x=X[label==1, 0], y=X[label==1, 1], marker='o', alpha=.8, c='g', label='Kind 2')
    plt.scatter(x=X[label==2, 0], y=X[label==2, 1], marker='s', alpha=.8, c='r', label='Kind 3')
    
    plt.legend(loc='upper right')
    
    plt.axis('equal')
    
    plt.plot()
    plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=12).fit(X)
print(kmeans.cluster_centers_)
pred_label = kmeans.predict(X)
kmeans_display(X, pred_label)
```
